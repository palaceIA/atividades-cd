```python
df = df.drop(columns=['Cabin'],axis=True)
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df['Age'] = df['Age'].astype(int)
```

```python
df['Sex']=df['Sex'].map({
    'male' : 0,
    'female' :1
})

df['Embarked']=df['Embarked'].map({
    'S' : 0,
    'C' : 1,
    'Q' : 2
})
```
